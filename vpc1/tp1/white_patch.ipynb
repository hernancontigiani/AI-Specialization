White Patch

```python
#Si queremos que las imágenes sean mostradas en una ventana emergente quitar el inline
#%matplotlib

# OpenCV-Python utiliza NumPy para el manejo de imágenes
import numpy as np
# cv2 es el módulo python para acceder a OpenCV 
import cv2 as cv
# Usamos las poderosas herramientas de graficación de matplotlib para mostrar imágenes, perfiles, histogramas, etc
import matplotlib.pyplot as plt
```

```python
# Cargar imagenes
img_1 = cv.imread('WP_B.png')
img_2 = cv.imread('WP_O.png')
img_3 = cv.imread('WP_R.png')
```

```python
# Muestro la imagen color
fig = plt.figure(figsize=(15,5))

ax1=plt.subplot(131)
ax1.imshow(img_1)

ax2=plt.subplot(132)
ax2.imshow(img_2)

ax3=plt.subplot(133)
ax3.imshow(img_3)
plt.show()
```

#### Algoritmo white_patch

```python
def white_patch(image):
    img = image.copy()
    color_max = [0] * 3
    color_max[0] = max(map(max,image[:,:,0]))
    color_max[1] = max(map(max,image[:,:,1]))
    color_max[2] = max(map(max,image[:,:,2]))
    #print(color_max)
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):            
            for k in range(image.shape[2]):
                img[i,j,k] = ((255)*image[i,j,k]) / color_max[k]
    return img
```

#### Balance de blancos en cada imagen

```python
img_mod_1 = white_patch(img_1);
np.clip(img_mod_1, 0, 255, out=img_mod_1)   # Clip trunca a lo que se le diga (0 a 255)
img_mod_1 = img_mod_1.astype('uint8')         # Convierto a 8 bits

img_mod_2 = white_patch(img_2);
np.clip(img_mod_2, 0, 255, out=img_mod_2)   # Clip trunca a lo que se le diga (0 a 255)
img_mod_2 = img_mod_2.astype('uint8')         # Convierto a 8 bits

img_mod_3 = white_patch(img_3);
np.clip(img_mod_3, 0, 255, out=img_mod_3)   # Clip trunca a lo que se le diga (0 a 255)
img_mod_3 = img_mod_3.astype('uint8')         # Convierto a 8 bits
```

```python
# Presentar resultados
fig = plt.figure(figsize=(20,10))
# Originales
ax1=plt.subplot(231)
ax1.imshow(img_1)

ax2=plt.subplot(232)
ax2.imshow(img_2)

ax3=plt.subplot(233)
ax3.imshow(img_3)

# Modificadas
ax1=plt.subplot(234)
ax1.imshow(img_mod_1)

ax2=plt.subplot(235)
ax2.imshow(img_mod_2)

ax3=plt.subplot(236)
ax3.imshow(img_mod_3)

plt.show()
```
