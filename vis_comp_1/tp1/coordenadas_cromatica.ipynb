Coordenadas cromáticas

```python
#Si queremos que las imágenes sean mostradas en una ventana emergente quitar el inline
#%matplotlib

# OpenCV-Python utiliza NumPy para el manejo de imágenes
import numpy as np
# cv2 es el módulo python para acceder a OpenCV 
import cv2 as cv
# Usamos las poderosas herramientas de graficación de matplotlib para mostrar imágenes, perfiles, histogramas, etc
import matplotlib.pyplot as plt
```

```python
# Cargar una imagen a color
img_1 = cv.imread('CoordCrom_1.png')
img_1 = cv.cvtColor(img_1, cv.COLOR_BGR2RGB)
img_2 = cv.imread('CoordCrom_2.png')
img_2 = cv.cvtColor(img_2, cv.COLOR_BGR2RGB)
img_3 = cv.imread('CoordCrom_3.png')
img_3 = cv.cvtColor(img_3, cv.COLOR_BGR2RGB)
```

```python
# Muestro la imagen color
plt.imshow(img_1)
plt.show()
print(img_1.shape)
```

#### Pasar a coordenadas cromáticas

```python
def coord_crom(image):
    img = image.copy()
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            r_g_b = image[i,j,0] + image[i,j, 1] + image[i,j, 2]
            if r_g_b == 0:
                continue
            for k in range(image.shape[2]):
                val = ((255)*image[i,j,k]) / r_g_b
                if val > 255:
                    val = 255
                img[i,j,k] = val
    return img
```

#### Resultado de pasar a coordenadas cromáticas

```python
img_crom_1 = coord_crom(img_1);
img_crom_1 = img_crom_1.astype('uint8')         # Convierto a 8 bits
plt.imshow(img_crom_1)
plt.show()
```

```python
img_mod_1 = coord_crom(img_1);
img_mod_1 = img_mod_1.astype('uint8')         # Convierto a 8 bits

img_mod_2 = coord_crom(img_2);
img_mod_2 = img_mod_2.astype('uint8')         # Convierto a 8 bits

img_mod_3 = coord_crom(img_3);
img_mod_3 = img_mod_3.astype('uint8')         # Convierto a 8 bits
```

```python
# Presentar resultados
fig = plt.figure(figsize=(20,10))
# Originales
ax1=plt.subplot(231)
ax1.imshow(img_1)

ax2=plt.subplot(232)
ax2.imshow(img_2)

ax3=plt.subplot(233)
ax3.imshow(img_3)

# Modificadas
ax1=plt.subplot(234)
ax1.imshow(img_mod_1)

ax2=plt.subplot(235)
ax2.imshow(img_mod_2)

ax3=plt.subplot(236)
ax3.imshow(img_mod_3)

plt.show()
```
